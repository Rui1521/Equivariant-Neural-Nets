### Eneregy Spectrum

```python
def TKE(preds):
    mean_flow = np.expand_dims(np.mean(preds, axis = 0), axis = 0)
    tur_preds = np.mean((preds - mean_flow)**2, axis = 0)
    tke = (tur_preds[:,0] + tur_preds[:,1])/2
    return tke

def tke2spectrum(tke):
    """Convert TKE field to spectrum"""
    sp = np.fft.fft2(tke)
    sp = np.fft.fftshift(sp)
    sp = np.real(sp*np.conjugate(sp))
    sp1D = radialProfile.azimuthalAverage(sp)
    return np.log10(sp1D) 
 
def spectrum_band(tensor):
    try:
        spec = np.array([tke2spectrum(TKE(tensor[i]*stds)) for i in range(tensor.shape[0])])
    except:
        spec = np.array([tke2spectrum(TKE(tensor[i][0]*stds)) for i in range(len(tensor))])
        return spec
    return np.mean(spec, axis = 0)#, np.std(spec, axis = 0)
```

```python
direc = ""
spec_rmse = []
for i in range(1,6):  
    preds = torch.load(direc + str(i) +".pt")["preds"]
    try:
        center = preds.shape[-1]//2
        preds = preds
        trues = torch.load(direc + "1.pt")["trues"]
        spec_rmse.append(np.sqrt(np.mean((spectrum_band(preds) - spectrum_band(trues))**2)))
    except:
        print("except")
        spec = []
        trues = torch.load(direc + "1.pt")["trues"]
        a = spectrum_band(preds)#)standardize(
        b = spectrum_band(trues)#)standardize(
        for i in range(len(a)):
            spec.append(np.sqrt(np.mean((a[i] - b[i])**2)))
        spec_rmse.append(np.mean(spec)) 
print(np.mean(spec_rmse), np.std(spec_rmse))
```
